Passo 1 – Criar o dicionário principal
1. Criar um dicionário que armazene diferentes distribuições e seus parâmetros

```python
distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": None},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": None}
}
```

```python
distribuicoes["Exponencial"]['params']['scale']=1
distribuicoes["Exponencial"]['params']['scale']
```

Passo 2 – Importar bibliotecas
2. Simular dados usando numpy.random

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Passo 3 – Gerar os dados

```python
np.random.normal(loc=0, scale=1)
```

```python
np.random.exponential(scale=1)
```

```python
np.random.uniform(low= -1, high= 1)
```

```python
amostra_normal_np = np.random.normal(loc=0, scale=1, size=1000)
```

```python
amostra_exp_np = np.random.exponential(scale=1, size=1000)
```

```python
amostra_uni_np = np.random.uniform(low= -1, high= 1, size=1000)
```

```python
list(amostra_normal_np)
list(amostra_exp_np)
list(amostra_uni_np)
```

```python
distribuicoes["Normal"]['dados']=amostra_normal_np
distribuicoes["Normal"]['dados']
```

```python
distribuicoes["Exponencial"]['dados']=amostra_exp_np
```

```python
distribuicoes["Uniforme"]['dados']=amostra_uni_np
```

Passo 4 – Calcular estatísticas básicas

```python
distribuicoes["Normal"]['estatísticas'] = {"média": np.mean(amostra_normal_np),"mediana": np.median(amostra_normal_np), "variância": np.var(amostra_normal_np),"desvio padrão": np.std(amostra_normal_np)}
```

```python
distribuicoes["Exponencial"]['estatísticas'] = {"média": np.mean(amostra_exp_np),"mediana": np.median(amostra_exp_np), "variância": np.var(amostra_exp_np),"desvio padrão": np.std(amostra_exp_np)}
```

```python
distribuicoes["Uniforme"]['estatísticas'] = {"média": np.mean(amostra_uni_np),"mediana": np.median(amostra_uni_np), "variância": np.var(amostra_uni_np),"desvio padrão": np.std(amostra_uni_np)}
```

Passo 5 – Visualizar os dados

```python
plt.hist(amostra_normal_np, bins=50)
```

```python
plt.hist(amostra_exp_np, bins=50)
```

```python
plt.hist(amostra_uni_np, bins=50)
```

Passo 7 – Adicionar novas distribuições

```python
distribuicoes["Poisson"] = {"params": {"lam": 4}, "dados": None, "estatisticas": None}
distribuicoes["Binomial"] = {"params": {"n": 9, "p": 0.5}, "dados": None, "estatisticas": None}
```

```python
np.random.poisson(lam=4)
```

```python
np.random.binomial(n=9, p=0.5)
```

```python
amostra_poisson_np = np.random.poisson(lam=4, size=1000)
```

```python
amostra_binomial_np = np.random.binomial(n=9, p=0.5, size=1000)
```

```python
list(amostra_poisson_np)
list(amostra_binomial_np)
```

```python
distribuicoes["Poisson"]['dados']=amostra_poisson_np
distribuicoes["Binomial"]['dados']=amostra_binomial_np
```

Passo 8 – Estatísticas adicionais

```python
distribuicoes["Poisson"]['estatistica'] = {"média": np.mean(amostra_poisson_np),"mediana": np.median(amostra_poisson_np), "variância": np.var(amostra_poisson_np),"desvio padrão": np.std(amostra_poisson_np), "min": np.min(amostra_poisson_np), "max": np.max(amostra_poisson_np), "quartis": np.percentile(amostra_poisson_np, [25, 50, 75])}
```

```python
distribuicoes["Binomial"]['estatistica'] = {"média": np.mean(amostra_binomial_np),"mediana": np.median(amostra_binomial_np), "variância": np.var(amostra_binomial_np),"desvio padrão": np.std(amostra_binomial_np), "min": np.min(amostra_binomial_np), "max": np.max(amostra_binomial_np), "quartis": np.percentile(amostra_binomial_np, [25, 50, 75])}
```

```python
distribuicoes["Normal"]['estatísticas']['min'] = np.min(distribuicoes["Normal"]["dados"])
distribuicoes["Normal"]['estatísticas']['max'] = np.max(distribuicoes["Normal"]["dados"])
distribuicoes["Normal"]['estatísticas']['quartis'] = np.percentile(distribuicoes["Normal"]["dados"], [25, 50, 75])
```

```python
distribuicoes["Exponencial"]['estatísticas']['min'] = np.min(distribuicoes["Exponencial"]["dados"])
distribuicoes["Exponencial"]['estatísticas']['max'] = np.max(distribuicoes["Exponencial"]["dados"])
distribuicoes["Exponencial"]['estatísticas']['quartis'] = np.percentile(distribuicoes["Exponencial"]["dados"], [25, 50, 75])
```

```python
distribuicoes["Uniforme"]['estatísticas']['min'] = np.min(distribuicoes["Uniforme"]["dados"])
distribuicoes["Uniforme"]['estatísticas']['max'] = np.max(distribuicoes["Uniforme"]["dados"])
distribuicoes["Uniforme"]['estatísticas']['quartis'] = np.percentile(distribuicoes["Uniforme"]["dados"], [25, 50, 75])
```

Passo 9 – Comparar distribuições graficamente

```python
plt.hist(amostra_normal_np, bins=50)
plt.hist(amostra_exp_np, bins=50)
plt.hist(amostra_uni_np, bins=50)
plt.hist(amostra_poisson_np, bins=50)
plt.hist(amostra_binomial_np, bins=50)
```

```python
plt.boxplot([amostra_normal_np, amostra_exp_np, amostra_uni_np, amostra_poisson_np, amostra_binomial_np])
plt.title('Comparação das Distribuições')
plt.ylabel('Valores')
plt.xlabel('Distribuições')
```

Passo 10 – Comparar estatísticas calculadas com valores teóricos

Normal

```python
valor_esp_n = distribuicoes["Normal"]["params"]["loc"]
plt.hist([valor_esp_n, distribuicoes["Normal"]['estatísticas']['média']])
plt.xlim([-0.01,0.1])
plt.show()
```

Exponencial

```python
valor_esp_exp = 1/distribuicoes["Exponencial"]["params"]["scale"]
plt.hist([valor_esp_exp, distribuicoes["Exponencial"]['estatísticas']['média']])
plt.xlim([0.9,1])
plt.show()
```

Uniforme

```python
valor_esp_uni = (distribuicoes["Uniforme"]["params"]["low"] + distribuicoes["Uniforme"]["params"]["high"])/2
plt.hist([valor_esp_uni, distribuicoes["Uniforme"]['estatísticas']['média']])
plt.xlim([0,0.02])
plt.show()
```

Poisson

```python
valor_esp_p = distribuicoes["Poisson"]["params"]["lam"]
plt.hist([valor_esp_p, distribuicoes["Poisson"]['estatistica']['média']])
plt.xlim([3.95,4])
plt.show()
```

Binomial

```python
valor_esp_b = (distribuicoes["Binomial"]["params"]["n"] * distribuicoes["Binomial"]["params"]["p"])
plt.hist([valor_esp_b, distribuicoes["Binomial"]['estatistica']['média']])

```
