## Exponencial
* Via inversão

```python
def exponencial(lam):
  u = np.random.uniform(low= 0, high= 1)
  x = (-1)*np.log(u)/(lam)
  return x
```

* Via geométrica

```python
def geometrica_inv(p):
  u = np.random.uniform(low= 0, high= 1)
  j = 0
  x = np.floor(np.log(1 - u)/np.log(1-p)) + 1

  return x

def aprox_exp_geo(lam, n):
  t = geometrica_inv(lam/n)
  x = t/n
  return x
```

## Normal
* Via aceitação-rejeição


```python
def normal_aceitacao_rejeicao():
  y = exponencial(1)
  u = np.random.uniform()
  while u > np.exp((-(y-1)**2)/2):
    y = exponencial(1)
    u = np.random.uniform()
  if u < np.exp((-(y-1)**2)/2):
    return y
```

* Norma padrão

```python
def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0

def normal_padrao():
  sinal = bernoulli(1/2)
  y = normal_aceitacao_rejeicao()
  if sinal == 1:
    return y
  else:
    return -y
```

* Transformação afim

```python
def transf_afim(l, e):
  z = normal_padrao()
  x = l + ((e**0.5)*z)
  return x
```
