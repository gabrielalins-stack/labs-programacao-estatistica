## Exercício 1

Aplique a técnica da variável antitética para estimar a integral

$$
I = \int_1^2 e^x \, dx.
$$

Compare o resultado com o método padrão (sem variável antitética).
Repita o experimento diversas vezes, estime a variância de cada método e apresente um histograma comparando as distribuições das estimativas obtidas com e sem o uso da variável antitética.

```python
import numpy as np
import matplotlib.pyplot as plt

def anti_tet(n):
  soma = 0
  for i in range(n):
    u = np.random.uniform()
    soma += (np.exp(u+1) + np.exp(2-u))

  return(soma/(2*n))

a = [anti_tet(1000) for i in range(1000)]
plt.hist(a, color = 'orchid', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
```

```python
def met_padrao(n):
  soma = 0
  for i in range(n):
    u = np.random.uniform()
    soma += np.exp(u+1)

  return(soma/n)

b = [met_padrao(1000) for i in range(1000)]
plt.hist(b, color = 'pink', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
plt.hist(a, color = 'orchid', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
```

## Exercício 2

Aplique a técnica da variável antitética para estimar o valor esperado de $f(X)$, onde $X \sim \text{Exp}(1)$ e $f(x) = e^{-x^2} $.

Compare o resultado com o método padrão (sem variável antitética).
Repita o experimento diversas vezes, estime a variância de cada método e apresente um histograma comparando as distribuições das estimativas obtidas com e sem o uso da variável antitética.

```python
def exponencial(lam):
  u = np.random.uniform(low= 0, high= 1)
  x = (-1)*np.log(u)/(lam)
  return x

def anti_tet2(n):
  soma = 0
  for i in range(n):
    u = np.random.uniform()
    x1 = -np.log(u)
    x2 = -np.log(1-u)
    soma += np.exp(-(x1**2)) + np.exp(-(x2**2))
  return(soma/(2*n))

c = [anti_tet2(1000) for i in range(1000)]
plt.hist(c, color = 'aquamarine', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
```

```python
def met_padrao2(n):
  soma = 0
  for i in range(n):
    x = exponencial(1)
    soma += np.exp(-(x**2))

  return(soma/n)

d = [met_padrao2(1000) for i in range(1000)]
plt.hist(d, color = 'teal', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
plt.hist(c, color = 'aquamarine', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
```

## Exercício 3

Aplique a técnica da variável antitética para estimar o valor esperado de $f(X)$, onde $X \sim \text{N}(1,2)$ e $f(x) = e^{-x^2+1} $.

Compare o resultado com o método padrão (sem variável antitética).
Repita o experimento diversas vezes, estime a variância de cada método e apresente um histograma comparando as distribuições das estimativas obtidas com e sem o uso da variável antitética.

```python
def normal_aceitacao_rejeicao():
  y = exponencial(1)
  u = np.random.uniform()
  while u > np.exp((-(y-1)**2)/2):
    y = exponencial(1)
    u = np.random.uniform()
  if u < np.exp((-(y-1)**2)/2):
    return y

def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0

def normal_padrao():
  sinal = bernoulli(1/2)
  y = normal_aceitacao_rejeicao()
  if sinal == 1:
    return y
  else:
    return -y

def transf_afim(l, e):
  z = normal_padrao()
  x = l + ((e**0.5)*z)
  return x

def anti_tet3(n):
  soma = 0
  for i in range(n):
    z = normal_padrao()
    soma += np.exp(-((2*z + 1)**2)+1) + np.exp(-((-2*z + 1)**2)+1)
  return(soma/(2*n))

e = [anti_tet3(1000) for i in range(1000)]
plt.hist(e, color = 'orchid', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
```

```python
def met_padrao3(n):
  soma = 0
  for i in range(n):
    z = normal_padrao()
    x = 2*z + 1
    soma += np.exp(-(x**2)+1)

  return(soma/n)

f = [met_padrao3(1000) for i in range(1000)]
plt.hist(f, color = 'pink', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
plt.hist(e, color = 'orchid', alpha = 0.7, bins=100, edgecolor = 'black', linewidth = 0.1)
```
