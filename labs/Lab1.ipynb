Exercício 1: Bernoullis

A - Gerador de Bernoulli

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0

amostra = []

for i in range(1000):
  a = bernoulli(0.3)
  amostra.append(a)

plt.hist(amostra)
```

B - Gerador de Uniforme via expansão binária

```python
def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0

def uniforme(k):
  u = 0
  for i in range(1,k):
    b_k = bernoulli(0.5)
    u += b_k/(2**i)
  return u

amostra1 = []

for i in range(1000):
  a = uniforme(20)
  amostra1.append(a)

amostra2 = np.random.uniform(low= 0, high= 1, size=1000)

plt.hist(amostra1)
plt.hist(amostra2)
```

Exercício 2: Binomial

```python
import time
start = time.time()
def binomial1(n, p):
  soma = 0
  for i in range(1, n):
    b_i = bernoulli(p)
    soma += b_i
  return soma

amostra_1 = []

for j in range(100000):
  a = binomial1(20, 0.3)
  amostra_1.append(a)

plt.hist(amostra_1)

end = time.time()
print("Tempo:", end - start)
```

```python
import time
start = time.time()
def binomial2(n, p):
  u = np.random.uniform(low= 0, high=1)
  p_i = (1-p)**n
  f = p_i
  i = 0
  while u > f:
    p_i = ((n-i)*p*p_i)/((i+1)*(1-p))
    f += p_i
    i += 1
  return i

amostra_2 = []

for j in range(100000):
  a = binomial2(20, 0.3)
  amostra_2.append(a)

plt.hist(amostra_2)

end = time.time()
print("Tempo:", end - start)
```
