##Bernoullis
* Gerador de Bernoulli

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0
```

* Gerador de Uniforme via expansão binária

```python
def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0

def uniforme(k):
  u = 0
  for i in range(1,k):
    b_k = bernoulli(0.5)
    u += b_k/(2**i)
  return u
```

## Binomial
* Soma de Bernoullis

```python
def binomial1(n, p):
  soma = 0
  for i in range(1, n):
    b_i = bernoulli(p)
    soma += b_i
  return soma
```

* Via recursão

```python
def binomial2(n, p):
  u = np.random.uniform(low= 0, high=1)
  p_i = (1-p)**n
  f = p_i
  i = 0
  while u > f:
    p_i = ((n-i)*p*p_i)/((i+1)*(1-p))
    f += p_i
    i += 1
  return i
```

## Poisson
* Via Inversão Recursiva

```python
def poisson_recursivo(lam):
  u = np.random.uniform(low= 0, high= 1)
  i = 0
  p0 = np.exp(-lam)
  f = p0
  while f < u:
    i += 1
    p0 = lam * p0 / i
    f += p0
  return i
```

* Via exponencial

```python
def exponencial(lam):
  u = np.random.uniform(low= 0, high= 1)
  x = (-1)*np.log(u)/(lam)
  return x

def poisson_via_exp(mu):
  tempo_total = 0
  contador = 0
  while tempo_total <= 1:
    tempo_total += exponencial(mu)
    contador += 1
  return contador - 1
```

## Geométrica
* Ingênua

```python
def bernoulli(p):
  u = np.random.uniform(low= 0, high= 1)
  if u <= p:
    return 1
  else:
    return 0

def geometrica_ingenua(p):
  i = 0
  j = 0
  while i == 0:
    i = bernoulli(p)
    j += 1
  return j+1
```

* Via inversão

```python
def geometrica_inv(p):
  u = np.random.uniform(low= 0, high= 1)
  j = 0
  x = np.floor(np.log(1 - u)/np.log(1-p)) + 1

  return x
```

## Binomial Negativa
* Soma de Bernoullis

```python
def neg_bin_ber(r, p):
  n = 0
  s = 0
  while s < r:
    b = bernoulli(p)
    n += 1
    if b == 1:
      s += 1
  return n
```

* Soma de geométricas

```python
def neg_bin_geo(r, p):
  soma_x = 0
  for i in range(1, int(r)):
    x_i = geometrica_inv(p)
    soma_x += x_i
  return soma_x
```

* Via inversão recursiva

```python
def neg_bin_inv(r, p):
  u = np.random.uniform()
  n = r
  p_n = p**r
  f = p_n
  while u > f:
    p_n = p_n*n*(1-p) / (n-r+1)
    n += 1
    f += p_n
  return n
```

## Hipergeométrica

```python
def hipergeom(N, K, n):
  pop = [1]*K + [0]*(N-K)
  for i in range(n):
    j = np.random.randint(i, N) #retorna um valor inteiro direto
    pop[j], pop[i] = pop[i], pop[j]
  return sum(pop[: n])
```
