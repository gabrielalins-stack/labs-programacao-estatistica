**Exercício 1**

Crie um vetor com 20 números inteiros aleatórios entre 0 e 100.

Mostre o vetor.

Calcule média, desvio-padrão e valor máximo.

```python
import numpy as np

vetor = np.random.randint(low=0, high=101, size=20)
print(vetor)

media = np.mean(vetor)
dp = np.std(vetor)
max = np.max(vetor)

print(media, dp, max)
```

**Exercício 2**

Gere uma matriz 5 X 5 com valores da distribuição normal padrão.

Calcule a soma de cada linha e de cada coluna.

Extraia a diagonal principal.

```python
matriz_5x5 = np.random.normal(loc=0, scale=1, size=(5,5)) #função que gera uma matriz de valores normais aleatórios com média 0 e variância 1 e tamnho específico
print(matriz_5x5)
```

```python
dim = np.array(matriz_5x5) #função que dimenciona a matriz
soma_col = np.sum(dim, axis=0) #função que soma as colunas
soma_lin = np.sum(dim, axis=1) #função que soma as linhas

print(soma_col, soma_lin)
```

```python
dig_principal = np.diagonal(matriz_5x5) #função que indica a diagonal principal da matriz
print(dig_principal)
```

**Exercício 3**

Implemente uma função que recebe dois vetores NumPy e retorna:

Produto escalar (np.dot).

Distância euclidiana entre eles (np.linalg.norm).

```python
def produto_escalar(vetor1, vetor2):
  produto_escalar = np.dot(vetor1, vetor2) #função de produto escalar
  return produto_escalar

vetor1 = np.random.normal(loc=0, scale=1, size=(5))
vetor2 = np.random.normal(loc=0, scale=1, size=(5))

print(produto_escalar(vetor1, vetor2))
```

```python
dist_euclid1 = np.linalg.norm(vetor1)
dist_euclid2 = np.linalg.norm(vetor2)

print(dist_euclid1, dist_euclid2)
```

**Exercício 4**

Crie uma matriz quadrada AER com valores inteiros aleatórios entre -5 e 5.

Verifique se A é invertível calculando det(A)(np.linalg.det).

Caso seja, calcule A^-1 (np.linalg.inv) e verifique se A*A^-1=I.

```python
matriz_a = np.random.randint(low=-5, high=6, size=(4,4)) #função que gera uma matriz com valores em um intervalo e tamanho específico
det_a = np.linalg.det(matriz_a) #função de determinante

print(det_a)
matriz_inv_a = np.linalg.inv(matriz_a) #função de inversa
mult = np.matmul(matriz_a, matriz_inv_a) #função de produto matricial
print(mult)
ident = np.identity(4)
print(ident)

if np.isclose(mult.all(), ident.all()):#função verifica se duas matrizes têm a forma próxima e todos os seus elementos são próximos
  print("True")
else:
  print("False")
```

```python
m = np.random.rand(3,3) #gera valores aleatórios, mas especificamos o tamanho
m_t = np.transpose(m) #função de transposta

b = (m + m_t) * 1/2
print(b)

autovalores, autovetores = np.linalg.eig(b)
print(autovalores, autovetores)
```

```python
mult1_1 = b @ autovetores[:,0] #@ multiplicação de matrizes
mult2_1 = autovetores[:,0] * autovalores[0]
print(mult1_1, mult2_1)

mult1_2 = b @ autovetores[:,1]
mult2_2 = autovetores[:,1] * autovalores[1]
print(mult1_2, mult2_2)

mult1_3 = b @ autovetores[:,0] #@ multiplicação de matrizes
mult2_3 = autovetores[:,0] * autovalores[0]
print(mult1_3, mult2_3)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('iris.csv')

plt.hist(df['sepal.length'], color = 'purple', alpha = 0.7, edgecolor='purple', linewidth=2)
plt.title('Tamanho da sépala X Quantidade')
plt.ylabel('Quantidade')
plt.xlabel('Tamanho da sépala')
```

```python
cores = {
    "Setosa": "aquamarine",
    "Versicolor": "teal",
    "Virginica": "turquoise"
}

for variety in df['variety'].unique():
  df_var = df[df['variety'] == variety]
  plt.scatter(df_var['sepal.length'], df_var['sepal.width'], label=variety, color = cores[variety],
              edgecolor='black', linewidth=0.5) #'label' é o nome da legenda

plt.title("Comprimento X largura da sépala")
plt.xlabel("Comprimento da sépala")
plt.ylabel("Largura da sépala")
plt.legend()
```
